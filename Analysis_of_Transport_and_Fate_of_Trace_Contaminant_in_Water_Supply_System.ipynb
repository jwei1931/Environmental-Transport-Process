Analyze the following amount of contaminant entering into the adjacent main pipe of water supply system from the natural source: 0.1 kg, 1 kg, 10 kg, respectively.

Occasionally discharge (0.1kg)


```
#Location 1,2
t1 = np.linspace(1, 10, num = 100)                #Assume that T starts from 1s since the graph goes to infinity
term1 = np.divide(1.443, np.sqrt(t1))
term2 = np.exp(np.multiply(-3.08, t1))
c1 = np.multiply(term1, term2)
a = c1.max()

print('The max conc at location 1 and 2 is ' + str(a) + 'kg/m3 when T=1s') 
```

```
plt.plot(t1, c1)
plt.axhline(y=a, color='red', linestyle='dashed')
plt.xlabel("T (s)")
plt.ylabel("c1 (kg/m^3)")
```

```
#Location 3
t3 = np.linspace(41250, 42100, num =1000)          #Based on the calculation, we can estimate the range where the peak is. 
term1_3 = np.divide(1.443, np.sqrt(t3))
term2_3 = np.subtract(50000, np.multiply(1.2, t3))
term3_3 = np.power(term2_3, 2)
term4_3 = np.divide(-term3_3, np.multiply(0.468, t3))
term5_3 = np.exp(term4_3)
c3 = np.multiply(term1_3, term5_3)
a = c3.max()
temp = ((42100-41250)/1000) * np.argmax(c3) + 41250
print('The max conc at location 3 is ' + str(a) + 'kg/m3 when T=' + str(temp) + 's') 
```

```
plt.plot(t3, c3)
plt.axhline(y=a, color='red', linestyle='dashed')
plt.xlabel("T (s)")
plt.ylabel("c3 (kg/m^3)")
```

```
#Reservoir1 (Location 4)
t4 = np.linspace(0, 7200, num =1000)  #Time at the reservoir
T = np.add(t4, t3)
term1_4 = np.exp(np.multiply(-0.000139, np.subtract(T, t3)))
term2_4 = np.subtract(1, term1_4)
term3_4 = np.divide(c3, 2)
c4 = np.multiply(term2_4, term3_4)
a = c4.max()
temp = ((7200)/1000) * np.argmax(c4) 
print('The max conc at location 4 is ' + str(a) + 'kg/m3 when t4=' + str(temp) + 's') 
```

```
plt.plot(t4, c4)
plt.axhline(y=a, color='red', linestyle='dashed')
plt.xlabel("t4 (s)")
plt.ylabel("c4 (kg/m^3)")
```
